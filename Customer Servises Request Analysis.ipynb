```python
#Import required libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#For statistical testing
from scipy import stats

#Display settings
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("C:/Users/shewa/OneDrive/Desktop/311-service-requests-nyc (1)/311_Service_Requests_from_2010_to_Present.csv")
```

# Frequency Plot of Null Values

```python
null_counts = df.isnull().sum()

plt.figure(figsize=(12,6))
null_counts[null_counts > 0].plot(kind='bar')
plt.title("Null Values per Column")
plt.show()
```

```python
df = df[df['Closed Date'].notnull()]
```

```python
df['Created Date'] = pd.to_datetime(df['Created Date'])
df['Closed Date'] = pd.to_datetime(df['Closed Date'])
```

```python
df = df[df['Closed Date'] >= df['Created Date']]
```

```python
df['Request_Closing_Time'] = df['Closed Date'] - df['Created Date']
```

```python
df['Request_Closing_Time'] = df['Request_Closing_Time'].dt.total_seconds()
```

```python
df['Request_Closing_Time'].describe()
```

```python
df[['Complaint Type', 'City']].isnull().sum()
```

```python
df['City'].fillna('Unknown City', inplace=True)
df['Complaint Type'].fillna('Unknown Complaint', inplace=True)
```

# Frequency Plot for City Complaints

```python
plt.figure(figsize=(12,6))
df['City'].value_counts().head(10).plot(kind='bar')
plt.title("Top Cities with Complaints")
plt.show()
```

# Scatter & Hexbin Plot (Brooklyn)

```python
brooklyn = df[df['City'] == 'BROOKLYN']

# Scatter
plt.figure(figsize=(6,6))
plt.scatter(brooklyn['Longitude'], brooklyn['Latitude'], alpha=0.3)
plt.title("Complaint Scatter Plot - Brooklyn")
plt.show()

# Hexbin
plt.figure(figsize=(6,6))
plt.hexbin(brooklyn['Longitude'], brooklyn['Latitude'], gridsize=50)
plt.title("Complaint Density - Brooklyn")
plt.show()
```

```python
# Select top complaint types
top_complaints = df['Complaint Type'].value_counts().head(5).index

groups = [df[df['Complaint Type'] == comp]['Request_Closing_Time'] for comp in top_complaints]
```

```python
stat, p = stats.kruskal(*groups)

print("Statistic:", stat)
print("P-value:", p)
```

```python
if p > 0.05:
    print("Fail to reject H0: All distributions are equal")
else:
    print("Reject H0: At least one distribution is different")
```
